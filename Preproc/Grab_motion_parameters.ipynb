```python
import pandas as pd
import numpy as np
import glob
import os
cwd = os.getcwd() 
print(cwd)
```

```python
confound_list = ['framewise_displacement', 'trans_x','trans_y','trans_z','rot_x','rot_y','rot_z']

base_dir = '/Users/yiyuwang/ImagingData/AVFP'


subjects = list(range(103, 127))

data_dir = base_dir + '/confounds'

res_dir = base_dir + '/fmri_info'
res_df = pd.DataFrame(subjects, columns = ['subject'])
# file_name = data_dir +'/sub-' + str(subjnum) + '_task-affvids_run-0' + str(r) + '_desc-confounds_regressors.tsv'
for confound in confound_list:
    for r in range(1,6): 
        a = []
        for subjnum in subjects:
            confounds_files = data_dir +'/sub-' + str(subjnum) + '_task-affvids_run-0' + str(r) + '_desc-confounds_regressors.tsv'
            #print(confounds_files)
            df = pd.read_csv(confounds_files, sep='\t')
            max_value = np.nanmax(df[confound].values)
            average_value = df[confound].mean()
            #print(max_value)
            a.append(max_value)

        res_df[confound + str(r)] = a
    
    

    
res_df.to_csv(res_dir + '/subject_motion_summary.csv',index=False) 
```
